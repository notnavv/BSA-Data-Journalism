# Data Cleaning
Standardize raw scraped data into clean CSVs for modeling.

```python
import pandas as pd

standings = pd.read_csv('../data_raw/standings.csv')
games     = pd.read_csv('../data_raw/games.csv')
advanced  = pd.read_csv('../data_raw/advanced_stats.csv')

print(standings.shape, games.shape, advanced.shape)
```

## 1. Clean Standings

```python
teams = standings.rename(columns={
    'W':    'wins',
    'L':    'losses',
    'W/L%': 'win_pct',
    'GB':   'games_back',
    'PS/G': 'pts_scored',
    'PA/G': 'pts_allowed',
    'SRS':  'srs',
})

# GB has '—' for the division leader — replace with 0
teams['games_back'] = teams['games_back'].replace('\u2014', 0).astype(float)

teams['games_played'] = teams['wins'] + teams['losses']
teams['point_diff']   = teams['pts_scored'] - teams['pts_allowed']

teams.head()
```

## 2. Merge Advanced Stats

```python
teams = teams.merge(advanced, on='Team', how='left')
teams.head()
```

## 3. Clean Games Log

```python
games['Date'] = pd.to_datetime(games['Date'])
games['margin'] = games['home_score'] - games['away_score']
games['home_win'] = (games['home_score'] > games['away_score']).astype(int)

print(games.dtypes)
games.head()
```

## 4. Save Clean Files

```python
teams.to_csv('../data_clean/teams_clean.csv', index=False)
games.to_csv('../data_clean/games_clean.csv', index=False)
print('Saved teams_clean.csv and games_clean.csv')
print(teams.columns.tolist())
```
